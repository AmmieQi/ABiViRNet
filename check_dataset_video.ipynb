## Check distribution of GT descriptions

```python
Y_val = []
for k,v in ds.extra_variables['val']['description'].iteritems():
    for s in v: 
        Y_val.append(s)
        
Y_test = []
for k,v in ds.extra_variables['test']['description'].iteritems():
    for s in v: 
        Y_test.append(s)

Y_train = ds.Y_train['description']
```

```python
names = ['train', 'val', 'test']
descriptions = [Y_train, Y_val, Y_test]
counts = [[], [], []]
for i,(n,d) in enumerate(zip(names,descriptions)):
    for s in d:
        counts[i].append(len(s.split(' ')))
    print "set", n
    print "mean: %0.4f words per sentence" % (np.sum(counts[i])/float(len(counts[i])))
    print "std_dev: %0.4f" % np.std(counts[i])
    print
```
